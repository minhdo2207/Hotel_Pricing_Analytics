```python
#function for extracting movie id 
def extract_id(url):
    sub1 = "https://www.imdb.com/title/"
    sub2 = "/?ref_"
    idx1 = url.index(sub1)
    idx2 = url.index(sub2)
    res = url[idx1 + len(sub1):idx2]
    return res
```

```python
movies_data['movie_id']=movies_data['url'].apply(extract_id)
first_column = movies_data.pop('movie_id')
movies_data.insert(0, 'movie_id', first_column)
movies_data.head(5)
```

```python
movies_data['count'] = pd.isnull(movies_data).sum(1)
movies_data= movies_data.sort_values(['count']).drop_duplicates(subset=['title'],keep='first').drop(['count'],axis=1)
movies_data.sort_index(inplace=True)
movies_data.head()
```

```python
def convert_num_rating_to_int(value):
    if isinstance(value, str):
        if 'K' in value:
            return int(float(value.replace('K', '')) * 1000)
        elif 'M' in value:
            return int(float(value.replace('M', '')) * 1000000)
        else:
            return int(float(value))
    elif isinstance(value, float):
        return int(value)
    else:
        return 0 # or raise an exception
#Replace NaN values with 0 
movies_data['num_user_review'].fillna(0,inplace=True)
movies_data['num_critic_review'].fillna(0,inplace= True)
#Convert num_rating,num_user_review,num_critic_review to integer
movies_data['num_rating'] = movies_data['num_rating'].apply(convert_num_rating_to_int)
movies_data['num_user_review'] = movies_data['num_user_review'].apply(convert_num_rating_to_int)
movies_data['num_critic_review'] = movies_data['num_critic_review'].apply(convert_num_rating_to_int)
```

```python
# Define a function to convert 'duration' to minutes
def convert_duration_to_minutes(duration_str):
    parts = duration_str.split()
    total_minutes = 0
    for part in parts:
        if 'h' in part:
            total_minutes += int(part.strip('h')) * 60
        elif 'm' in part:
            total_minutes += int(part.strip('m'))
    return total_minutes
# Apply the function to convert 'duration' values to minutes
movies_data['duration'] = movies_data['duration'].apply(convert_duration_to_minutes)
```

```python
# oscar count
# Define a function to extract the number of Oscars won
def extract_oscar_count(oscar_str):
    if isinstance(oscar_str, str) and 'oscar' in oscar_str.lower() and 'won' in oscar_str.lower():
        # Extract the number of Oscars won
        words = oscar_str.split()
        for i in range(len(words)):
            if words[i].isdigit() and words[i + 1].lower() == 'oscar' and words[i - 1].lower() == 'won':
                return int(words[i])
    # If not 'won' and 'oscar' or no number found, return 0
    return 0

# Apply the function to convert 'oscar' values to the number of Oscars won
movies_data['oscar'] = movies_data['oscar'].apply(extract_oscar_count)
```

## Extract wins and nominations and create new columns

```python

wins = movies_data['win_and_nomination'].str.extract(r'(\d+) wins?')
nominations = movies_data['win_and_nomination'].str.extract(r'(\d+) nominations?')

# Fill missing values with 0
wins.fillna(0, inplace=True)
nominations.fillna(0, inplace=True)

# Add the extracted columns to the DataFrame
movies_data['win'] = wins
movies_data['nomination'] = nominations
movies_data[['title','win_and_nomination','win','nomination']].head()
```

```python
# create new columns for each genre
movies_data['genres'] = movies_data['genres'].fillna('')
movies_data['genres'] = movies_data['genres'].str.split(',')
unique_genres = set(genre for genres in movies_data['genres'] for genre in genres)
for genre in unique_genres:
    movies_data[genre] = movies_data['genres'].apply(lambda x: 1 if genre in x else 0)
```

## Impute the "metascore" column

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

# Create a copy of the dataframe for imputation
movies_data_imputed = movies_data.copy()

# Select the column to impute
column_to_impute = ['metascore']

# Reshape the features for imputation
X = movies_data_imputed[column_to_impute].values.reshape(-1, 1)

# Initialize the imputer with a predictive model (e.g., RandomForestRegressor)
imputer = IterativeImputer(estimator=RandomForestRegressor(), max_iter=10, random_state=0)

# Fit and transform the data
movies_data_imputed[column_to_impute] = imputer.fit_transform(X)
movies_data['metascore']=movies_data_imputed['metascore']
movies_data['metascore'] = movies_data['metascore'].astype('int')
```
